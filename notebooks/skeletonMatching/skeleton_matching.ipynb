```python
import networkx as nx
import numpy as np
from skeletontypes import JointType, BoneType
```

```python
def create_quadruped():
    """
    @brief Creates a quadruped skeleton.
    @returns Skeleton type, X
    """

    quad_like = nx.Graph()
    quad_like.add_node(0, pos=np.asarray([0.25, 0], dtype=float), type=JointType.LIMB)
    quad_like.add_node(1, pos=np.asarray([0.50, 0], dtype=float), type=JointType.LIMB)
    quad_like.add_node(2, pos=np.asarray([0.62, 0], dtype=float), type=JointType.LIMB)
    quad_like.add_node(3, pos=np.asarray([0.93, 0], dtype=float), type=JointType.LIMB)
    quad_like.add_node(4, pos=np.asarray([0.37, 1], dtype=float), type=JointType.BODY)
    quad_like.add_node(5, pos=np.asarray([0.75, 1], dtype=float), type=JointType.BODY)
    quad_like.add_node(6, pos=np.asarray([1, 0.80], dtype=float), type=JointType.BODY)
    quad_like.add_node(7, pos=np.asarray([0, 1], dtype=float), type=JointType.HEAD)

    quad_like.add_edge(0, 4, type=BoneType.MIXED)
    quad_like.add_edge(1, 4, type=BoneType.MIXED)
    quad_like.add_edge(2, 5, type=BoneType.MIXED)
    quad_like.add_edge(3, 5, type=BoneType.MIXED)
    quad_like.add_edge(4, 5, type=BoneType.BODY)
    quad_like.add_edge(5, 6, type=BoneType.BODY)
    quad_like.add_edge(4, 7, type=BoneType.MIXED)

    return quad_like
```

```python
def draw(graph):
    """
    @brief Helper function for drawing the given graph
    @param graph Graph that will be drawn
    @returns None
    """
    colors = ['green', 'red', 'green', 'blue', 'grey']
    color_map = []
    for nid, attr in graph.nodes(data=True):
        color_map.append(colors[int(attr['type'])])

    x = nx.get_node_attributes(graph, 'pos')
    nx.draw(graph, x, node_color=color_map, alpha=0.5)
```

```python
quadruped = create_quadruped()
draw(quadruped)
```
