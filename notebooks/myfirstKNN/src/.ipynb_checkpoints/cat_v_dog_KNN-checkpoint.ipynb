```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("../data/cat/3841.png")
```

```python
target_size = (32,32)
```

```python
resized_img = cv.resize(img, target_size)
```

```python
f, axarr = plt.subplots(1,2)
axarr[0].imshow(img)
axarr[0].set_title("original")
axarr[1].imshow(resized_img)
axarr[1].set_title("resized")
plt.show()
```
